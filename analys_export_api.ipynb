```python
import pandas as pd
import requests
```

```python
base_url = 'https://data-charts-api.hexlet.app'

visits_url = f'{base_url}/visits?begin=2023-03-01&end=2023-09-01'
registrations_url = f'{base_url}/registrations?begin=2023-03-01&end=2023-09-01'

visits_response = requests.get(visits_url)
registrations_response = requests.get(registrations_url)
```

```python
visits_response.status_code, registrations_response.status_code
```

```python
visits_api = pd.DataFrame(visits_response.json())
registrations_api = pd.DataFrame(registrations_response.json())
```
